```python
# run RSM eigen decomposition analysis

from numpy import linalg as LA

layer_idx = 9
epoch = 'best'
PATH_figures = '/Users/shahab/Mila/Results/DeepMouse/RSM/Mouse_vs_Models/' + area + '/'
ModelType = 'eigendecomp_CPC_simmousenet_pooled_wsplit_fscr_'
StimType = 'NM_full_'
ImagingDepth = '175'
CreLine = 'Cux2-CreERT2'
thisFigureName = PATH_figures + ModelType + StimType + ImagingDepth + CreLine + 'epoch_' + str(epoch) + '_layer ' + str(layer_idx)+'.svg'

RSM_Allen_filldiag = np.mean(all_RSM_Allen[:,:,:],axis=2)
# np.fill_diagonal(RSM_Allen_filldiag,'nan')
plt.imshow(RSM_Allen_filldiag)
plt.colorbar()
plt.title('brain RSM',fontsize=15)
plt.show()

RSM_one_layer = all_RSM_CPC[list(all_RSM_CPC.keys())[layer_idx]]
if model_type != 'I3D':
    subsample_rate = int(RSM_one_layer.shape[0]/all_RSM_Allen.shape[0])
    RSM_one_layer = RSM_one_layer[::subsample_rate,::subsample_rate]
else:
    RSM_one_layer = sp.signal.resample(sp.signal.resample(RSM_one_layer,num=all_RSM_Allen.shape[0],axis=0),num=all_RSM_Allen.shape[0],axis=1)
#         print(subsample_rate,conv1.shape[0])


plt.imshow(RSM_one_layer)
plt.colorbar()
plt.title('model RSM',fontsize=15)
plt.show()

w, v = LA.eig(RSM_one_layer)
idx_w_sort = np.argsort(w)

scale = np.empty((all_RSM_Allen.shape[2],w.shape[0]))

for i in range(all_RSM_Allen.shape[2]):
    w_v1, v_v1 = LA.eig(all_RSM_Allen[:,:,i])
    idx_wv1_sort = np.argsort(w_v1)
    
    counter = 0
    for eigcounter in idx_wv1_sort:#range(w_v1.shape[0]):
        scale[i,counter] = LA.norm(np.matmul(all_RSM_Allen[:,:,i],v[:,eigcounter]))
        counter += 1
    plt.plot(scale[i,:],'.k',alpha=.05)
    plt.plot(w_v1[idx_wv1_sort],'.r',alpha=.05)
    
    
plt.legend(['Brain ($\lambda_i$)','Model'],fontsize=12)
plt.xlabel('eigenvector index',fontsize=15)
plt.ylabel('Scaling',fontsize=15)
plt.title('CPC layer '+str(layer_idx),fontsize=15)
plt.savefig(thisFigureName,dpi=300)
plt.show()


u, s, vh = LA.svd(np.mean(all_RSM_Allen[:,:,:],axis=2))
# u, s, vh = LA.svd(RSM_one_layer)
# w, v = LA.eig(RSM_one_layer)
    
# show the first eigen matrix for brain RSM   
ModelType = 'eigendecomp_Vp_'
thisFigureName = PATH_figures + ModelType + StimType + ImagingDepth + CreLine +'.svg'
eigenmatrix = s[0]*np.outer(u[:,0],np.reshape(u[:,0],(1,100)))
plt.imshow(eigenmatrix)
plt.colorbar()
plt.title('brain RSM eigen matrix',fontsize=15)
plt.savefig(thisFigureName,dpi=300)
plt.show()

```
